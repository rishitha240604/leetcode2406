```python
import zipfile
import os

# Define paths
zip_file_name = 'archive (5).zip'
extract_dir = './plantvillage dataset/'

# Extract the ZIP file
with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# List the extracted directories
categories = os.listdir(extract_dir)
print(categories)
```
